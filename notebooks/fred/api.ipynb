```python
%reload_ext autoreload
%autoreload 2

import asyncio
from lib.clients import FredClient
from lib.utils import print_json_vertical
```

```python
def print_observations_vertical(observations):
    for obs in observations:
        print(f"realtime_start: {obs.realtime_start}")
        print(f"realtime_end:   {obs.realtime_end}")
        print(f"date:           {obs.date}")
        print(f"value:          {obs.value}\n")
```

## /series

```python

async def show_raw_series(series_id='GDP'):
    async with FredClient() as client:
        raw = await client._get('/series', {'series_id': series_id})
    print_json_vertical(raw)

await show_raw_series()
```

```python
async def demo():
    async with FredClient() as client:
        seriess = await client.get_series("GDP")
        print(seriess.seriess[0].title)
        observations = await client.get_series_observations("GDP", limit=100)
        print_observations_vertical(observations.observations)

await demo()
```
